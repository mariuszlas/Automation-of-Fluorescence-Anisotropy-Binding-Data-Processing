```python
skiprows = 0

def read_in_B(file): 
    reader = csv.reader(open(file, 'r'), delimiter=',') 
    
    all_data = [] 

    for row in reader:
        all_data.append(row)   # Create list of list with all data in the csv file
        
        if row != []:   # Blank rows cannot be indexed 
            if re.search(r"Exported with EnVision Workstation", row[0]) != None:   # Find line number of the last row
                total_lines = reader.line_num
    
        if row != [] and len(row) >= 2:   
                                           
            if re.findall(r"Plate", row[0]) == ['Plate']  and re.findall(r"Barcode", row[1]) == ['Barcode']:   # find line number with the beggining of the data
                global skiprows
                skiprows = reader.line_num - 1
                
    end_of_data = 0

    while end_of_data == 0:   
        for index, item in enumerate(all_data):
            if item == [] and index > skiprows:   # find the index of the first blank row after the data table
                end_of_data = end_of_data + index
                break

    skipfooter = total_lines - end_of_data   # calculate skipfooter parameter
    
    df = pd.read_csv(file, sep=',', index_col=0, engine='python', skiprows=skiprows, skipfooter=skipfooter, encoding='utf-8')
    
    return df
```
