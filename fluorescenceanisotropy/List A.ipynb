```python
skiprows = 0

def read_in(file): 
     
    reader = csv.reader(open(file, 'r'), delimiter=',')   # converts each line into a list
    
    for row in reader:
        if row != []:   # some rows are blank and so are converted to empty lists which cannot be indexed
            if re.findall(r"PlateNumber", row[0]) == ['PlateNumber'] and re.findall(r"PlateRepeat", row[1]) == ['PlateRepeat']:   # find the line where data to be imported starts
                global skiprows
                skiprows = reader.line_num - 1
                
            if re.search(r"Exported with EnVision Workstation", row[0]) != None:   # find total number of lines in the file
                total_lines = reader.line_num    
                
        if row == [] and reader.line_num > skiprows:   # find the line where the data to be imported ends by finding the closest blank row after data headers
            end_of_data = reader.line_num

    skipfooter = total_lines - end_of_data   
    
    df = pd.read_csv(file, sep=',', index_col=0, engine='python', skiprows=skiprows, skipfooter=skipfooter, encoding='utf-8')
    
    return df
```
